```python
import catppuccin
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use(catppuccin.PALETTE.latte.identifier)
```

```python
df = pd.read_excel("TSP-Resultados.xlsx")
```

```python
df
```

```python
tempo = df.iloc[1:,[0,1,3,8]]
tempo
```

```python
x = tempo['Unnamed: 1']
y1 = tempo['Twice-around-the-tree']
y2 = tempo['Christofides']

# Criando o gráfico
plt.figure(figsize=(12, 8))
plt.plot(x, y1, label='Twice-around-the-tree', marker='o')
plt.plot(x, y2, label='Christofides', marker='s')

plt.yticks(range(0, 1900, 100))
plt.xticks(range(0, 19000, 1000))

# Adicionando título e rótulos
plt.title('Tempos de Execução, Twice around the tree e Christofides')
plt.xlabel('Tamanho da Instância')
plt.ylabel('Tempo (s)')
plt.legend()
plt.grid(True)
```

```python
qual = df.iloc[1:74, [0,1,5,10]]
qual
```

```python
x = qual['Unnamed: 1']
y1 = qual['Unnamed: 5']
y2 = qual['Unnamed: 10']

# Criando o gráfico
plt.figure(figsize=(12, 8))
plt.plot(x, y1, label='Twice-around-the-tree', marker = "o", linewidth=1, color = 'orange')
plt.plot(x, y2, label='Christofides', marker= 's', linewidth=1)


med_twice = qual['Unnamed: 10'].mean()
med_chris = qual['Unnamed: 5'].mean()
# plt.ylim(0, 100)
# plt.xlim(0,500)

plt.axhline(med_chris, color='red', linestyle='--', label='Média Twice-around-the-tree')
plt.axhline(med_twice, color='green', linestyle='--', label='Média Christofides')

plt.yticks([1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, round(med_twice, 2), round(med_chris, 2)])

# Adicionando título e rótulos
plt.title('Qualidade das Respostas, Twice around the tree e Christofides')
plt.xlabel('Tamanho da Instância')
plt.ylabel('Qualidade (Resposta / Ótimo)')
plt.legend()
```
